```python
symbol_list = os.listdir(tick_data_path)
```

```python
for symbol in tqdm(symbol_list):
    
    contract_list = os.listdir(tick_data_path + '/' + symbol)
    
    for contract in tqdm(contract_list):
        
        df = pd.read_csv(tick_data_path + '/' + symbol + '/' + contract, index_col='trading_date', 
                         parse_dates=True, usecols=['Contract_Code', 'trading_date','a1_v', 'b1', 'b1_v'])
        
        df.columns = ['Contract_Code', 'b1', 'a1_v', 'b1_v']
        contract_code = df['Contract_Code'].values[0]
        
        new_df = df.loc[:, 'a1_v':'b1_v'].groupby(['trading_date']).sum()
        new_df.to_csv('../data/interim/bid_and_ask_volume/' + symbol + '/' + contract_code + '_daily_bid_and_ask.csv')
        
        time.sleep(0.01)
```
