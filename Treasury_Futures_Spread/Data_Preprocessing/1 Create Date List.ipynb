```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import os
import glob
```

```python
treasury_data_path = r'..\data\raw\daily_data'
```

```python
treasury_list = ['T', 'TF', 'TS']
name_list = []
# collect all contract names from the directory
for treasury in treasury_list:
    oslist = os.listdir(treasury_data_path + '/' + treasury)
    name_list = name_list + [x.split('.')[0] for x in oslist]
```

```python
name_list
```

```python
wind_trading_day = pd.read_csv(r'..\data\raw\meta_data\wind_trading_day.csv', index_col='Trading_Day', parse_dates=True)
```

```python
def convert_to_datetime(name, shift=None):# shift from intention day, default is None
    
    date = name[-4:]
    year = '20' + date[:2]
    month = date[-2:]
    
    first_natural_day = pd.to_datetime(year+month+'01', format='%Y%m%d')
    
    if shift is None: # return first_intention_day, first_notice_day and delivery_day
        delivery_day = wind_trading_day.loc[first_natural_day:].index[0]
        first_notice_day = wind_trading_day.loc[:delivery_day].index[-2]
        first_intention_day = wind_trading_day.loc[:delivery_day].index[-3]
        return [first_intention_day, first_notice_day, delivery_day]
    
    else: # return three consecutive days depending on shift
        trading_day = pd.Series(list(wind_trading_day.index), index=range(len(wind_trading_day)))
        wind_trading_day_ = pd.Series(range(len(wind_trading_day)), index=wind_trading_day.index)
        idx = wind_trading_day_.loc[first_natural_day:][0]
        idx_ = idx + shift
        third_day = trading_day[idx_]
        second_day = trading_day[idx_ - 1]
        first_day = trading_day[idx_ - 2]
        return [first_day, second_day, third_day]
```

```python
# return intention day, first_notice_day and delivery_day
date_matrix = [convert_to_datetime(x) for x in name_list]
```

```python
date_df = pd.DataFrame(date_matrix, index=name_list, columns=['intention_day', 'notice_day', 'delivery_day'])
#date_df = pd.DataFrame(date_matrix, index=name_list, columns=['first_day', 'second_day', 'third_day'])
date_df
```
